```python
from Bio import SeqIO
import os,sys
import pandas as pd
```

```python
import numpy as np
```

```python
def map_mutation(input_fasta,output):
    seq_records = SeqIO.parse(input_fasta, 'fasta')
    df = pd.DataFrame(columns=['seq_id','location','mutate','ref'])
    n=0

    refseq_record =next(seq_records)
    for seq_record in seq_records:
        for i in range(0, len(refseq_record)):
            nt1 = refseq_record[i]
            nt2 = seq_record[i]
            if nt1 != nt2:
                n=n+1
                df.loc[n, ['seq_id']] = seq_record.id
                df.loc[n, ['location']] = i+1
                df.loc[n, ['mutate']] = nt2
                df.loc[n, ['ref']] = nt1
                #print(seq_record.id, i+1, nt2, nt1)
                
   
    df.to_csv(output, index=False)
    return(df)          
```

```python
S_pro_mut=map_mutation("human_CoV/human_gisaid_clean_astr/S_protein.fasta","output.csv")
```

```python
def fx(x):
    if np.isnan(x['location_y']):
        return(str(x["mutate_x"])+str(x['column']))
    else:
        return(str(x['ref_y'])+str(x['location_y'])+str(x['mutate_y']))
    
  
```

```python
new_df['Temp']=new_df.apply(lambda x : fx(x),axis=1)  
```

```python
new_df
```

```python
def lolipop_mut(pro_mut,nuc_mut):
    nuc_mut['location2']=nuc_mut['location']/3
    decimals = 0    
    nuc_mut['column'] = nuc_mut['location2'].apply(lambda x: round(x, decimals))
    new_df=pd.merge(nuc_mut,pro_mut,how='left', left_on=['seq_id','column'],right_on=['seq_id','location'])

    new_df['Temp']=new_df.apply(lambda x : fx(x),axis=1)
    return(new_df)
```

```python
S_mutmap=lolipop_mut(S_pro_mut,S_nuc_mut)
```
