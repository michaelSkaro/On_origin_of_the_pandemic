```python
from Bio import SeqIO
import os,sys
import pandas as pd
import phylopandas as ph
import numpy as np
```

```python
## convert fasta to df
def fasta2df(fasta):
    df = ph.read_fasta(fasta)
    df = df.filter(['id','sequence'], axis=1)
    return(df)
    
    
```

```python
df1=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/orf1ab_al.fasta")
df2=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/S_al.fasta")
df3=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/orf3a_al.fasta")
df4=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/CoV_E_nuc_al.fasta")
df5=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/M_al.fasta")
df6=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/orf6_al.fasta")
df7=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/orf7a_al.fasta")
df8=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/CoV_orf7b_nuc_al.fasta")
df9=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/orf8_al.fasta")
df10=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/N_al.fasta")
df11=fasta2df("Alignment_0307/nuc_gene/nuc_alignment0315/CoV_orf10_nuc_al.fasta")
```

```python
## merge all df
dfs = [df1, df2, df3,df4,df5,df6,df7,df8,df9,df10,df11]
dfs = [x.set_index('id') for x in dfs]

df = pd.concat(dfs, axis=1, keys=range(1, len(dfs) + 1))
df.columns = df.columns.map('{0[1]}{0[0]}'.format)
df = df.reset_index()
```

```python
df
```

```python
df['WGS']=df['sequence1'].fillna('')+df['sequence2'].fillna('')+df['sequence3'].fillna('')+df['sequence4'].fillna('')+df['sequence5'].fillna('')+df['sequence6'].fillna('')+df['sequence7'].fillna('')+df['sequence8'].fillna('')+df['sequence9'].fillna('')+df['sequence10'].fillna('')+df['sequence11'].fillna('')
    
WGS = df.filter(['index','WGS'], axis=1)
```

```python
WGS
```

```python
## Convert WGS to fasta
WGS.to_csv("temp.tab", sep="\t",index = False,header=False)
fasta = SeqIO.parse("temp.tab", "tab")
SeqIO.write(fasta,"Alignment_0307/nuc_gene/nuc_alignment0315/CoV_WGS_nuc.fasta","fasta")
os.remove("temp.tab")
```
